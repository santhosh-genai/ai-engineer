```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import mean_squared_error, r2_score
```

```python
data = pd.read_csv('C:\\Users\\2276038\\Desktop\\learning\\ai-engineer\\python\\assignments\\Assignment_week21\\sales_data.csv')
X = data[['Advertising_Spend']]  # Input feature - Independent variable (input)
Y = data['Total Amount']    # Target variable - Dependent variable (output)
print(data.describe())
```

```python
plt.scatter(X, Y , color='blue')
plt.title('Advertising Spend vs Total Amount (Sales)')
plt.xlabel('Advertising Spend') 
plt.ylabel('Total Amount (Sales)')
plt.title('Scatter Plot: Advertising Spend vs Total Amount (Sales)')
plt.show()
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)

model = LinearRegression()
model.fit(X_train, Y_train)

Y_pred = model.predict(X_test)

print("mse:", mean_squared_error(Y_test, Y_pred))
print("rmse:", np.sqrt(mean_squared_error(Y_test, Y_pred)))
print("r2:", r2_score(Y_test, Y_pred))

plt.scatter(X_test, Y_test, color='blue', label='Actual Values')
plt.scatter(X_test, Y_pred, color='red', label='Predicted Values')
plt.plot(X_test, Y_pred, color='green', label='Regression Line')
plt.xlabel('Advertising Spend')
plt.ylabel('Total Amount (Sales)')
plt.title('Simple Linear Regression: Actual vs Predicted Values')
plt.legend()
plt.show()
```

```python
poly = PolynomialFeatures(degree=2)
# Transform X_train and X_test (use the earlier train/test split)
X_train_poly = poly.fit_transform(X_train)
X_test_poly = poly.transform(X_test)
model = LinearRegression()
model.fit(X_train_poly, Y_train)
Y_pred = model.predict(X_test_poly)
# ensure 1D arrays for metric calculations and plotting
Y_pred = np.ravel(Y_pred)
print("mse:", mean_squared_error(Y_test, Y_pred))
print("rmse:", np.sqrt(mean_squared_error(Y_test, Y_pred)))
print("r2:", r2_score(Y_test, Y_pred))

# smooth curve for polynomial regression: predict on a dense X grid
X_range = np.linspace(X.min()[0], X.max()[0], 200).reshape(-1, 1)
X_range_poly = poly.transform(X_range)
y_range_pred = model.predict(X_range_poly)

plt.scatter(X_test, Y_test, color='blue', label='Actual Values')
plt.scatter(X_test, Y_pred, color='red', label='Predicted Values')
plt.plot(X_range, y_range_pred, color='green', label='Polynomial Regression Curve')
plt.xlabel('Advertising Spend')
plt.ylabel('Total Amount (Sales)')
plt.title('Polynomial Regression (Degree 2): Actual vs Predicted Values')
plt.legend()
plt.show()
```
